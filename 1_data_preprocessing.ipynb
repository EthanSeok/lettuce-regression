```python
FILE_DIRECTORY = '/content/gdrive/My Drive/Colab Notebooks/PartA/data/GroundTruth.json'
```

```python
with open(FILE_DIRECTORY) as data_file:    
    JSON_data = json.load(data_file)
```

```python
JSON_data.keys()
```

```python
image_index = np.array(list(JSON_data['Measurements'].keys()))
image_index
```

```python
shoot_fw = []
shoot_dw = []
height = []
diameter = []
leafarea = []
variety = []
for _ in JSON_data['Measurements'].keys():
    shoot_fw.append(JSON_data['Measurements'][_]['FreshWeightShoot'])
    shoot_dw.append(JSON_data['Measurements'][_]['DryWeightShoot'])
    height.append(JSON_data['Measurements'][_]['Height'])
    diameter.append(JSON_data['Measurements'][_]['Diameter'])
    leafarea.append(JSON_data['Measurements'][_]['LeafArea'])
    variety.append(JSON_data['Measurements'][_]['Variety'])
```

```python
output_df = pd.DataFrame(np.array([shoot_fw, shoot_dw, height, diameter, leafarea]).T, index=image_index, columns=['shoot_fw', 'shoot_dw', 'height', 'diameter', 'leafarea'])
```

```python
output_df.to_csv('/content/gdrive/My Drive/Colab Notebooks/PartA/results/output_df.csv')
```

```python
variety_df = pd.DataFrame(np.array([shoot_fw, shoot_dw, height, diameter, leafarea, variety]).T, index=image_index, columns=['shoot_fw', 'shoot_dw', 'height', 'diameter', 'leafarea', 'variety'])
```

```python
variety_df.to_csv('/content/gdrive/My Drive/Colab Notebooks/PartA/results/variety_df.csv')
```
